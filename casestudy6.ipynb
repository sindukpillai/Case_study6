```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv('C:\\Users\\HP\Desktop\\Add_sales_data_2018.csv')
data.head()
```

```python
# all adds have strong positive correlation with profit
corrmatrix=data.corr()
plt.subplots(figsize=(10,4))
sns.heatmap(corrmatrix,vmin=.8,vmax=1,annot=True,linewidth=.2,cmap='YlGnBu')
data.columns
```

```python
#regression model to predict the profit from the given data
y=data['Profit(in Thousands)']
x=data.drop(['Date','Add2(in Thousands)','Profit(in Thousands)'],axis=1)

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,
                                               test_size=.25)
from sklearn import linear_model
lr=linear_model.LinearRegression()
model=lr.fit(x_train,y_train)
predictions=model.predict(x_test)
print(y_test)
print(predictions)
```
